```python
import scipy.linalg as la
import matplotlib.pyplot as plt
from math import sin, exp
import numpy as np

abscissa = np.arange(1., 15.1, .1)

def f(x):
    return sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)

def p(x, W, dim):
    return sum([W[p] * x ** p for p in range(dim + 1)])

def build_matrix(x_coords, dim):
    rows = []
    for l in range(dim + 1):
        row = []
        for p in range(dim + 1):
            row.append(x_coords[l] ** p)
        rows.append(row)
    
    return rows

def approximate(x_match, dim):
    B = [f(x) for x in x_match]
    A = build_matrix(x_match, dim)
    W = la.solve(A, B)
    return W
```

# Полином первой степени

```python
n = 1
x_match = [1., 15.]
W = approximate(x_match, n)
plt.plot(abscissa, [f(x) for x in abscissa], abscissa, [p(x, W, n) for x in abscissa])
```
